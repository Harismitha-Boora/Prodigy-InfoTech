# import require module

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

# import dataset

```python
df=pd.read_csv("bank-marketing prodigy task 3.csv")
```

```python
df
```

# processing the data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()
```

```python
df["job"]=l.fit_transform(df.job)
df["marital"]=l.fit_transform(df.marital)
df["education"]=l.fit_transform(df.education)
df["default"]=l.fit_transform(df.default)
df["housing"]=l.fit_transform(df.housing)
df["loan"]=l.fit_transform(df.loan)
df["contact"]=l.fit_transform(df.contact)
df["month"]=l.fit_transform(df.month)
df["poutcome"]=l.fit_transform(df.poutcome)
df["y"]=l.fit_transform(df.y)
```

```python
df
```

# Splitting the dataframe into independent and dependent variable

```python
x=df[["age","job","marital","education","default","balance","housing","loan","contact","day","month","duration","campaign","pdays","previous","poutcome"]]
y=df[["y"]]
```

# splitting the data into train and test

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=2500,random_state=1)
```

```python
train_x
```

```python
train_y
```

```python
d=DecisionTreeClassifier()
import warnings
warnings.filterwarnings("ignore")
```

# Training the model

```python
d.fit(train_x,train_y)
```

# Predictting the model

```python
d.predict([[55,4,1,3,0,1504,1,0,0,22,8,338,2,-1,0,3]])
```

```python
d.predict([[24,1,2,1,0,171,1,0,2,21,8,151,2,-1,0,3]])
```

# Accuracy of the model

```python
d.score(train_x,train_y)
```
